```python
import csv
with open('appart.csv','w',encoding='utf-8',newline='') as f:
    writer=csv.writer(f)
    writer.writerow(['title','location','area','price' ,'Number of rooms' ,'furnished or not'])
    
```

```python
row=[]
page_num=0
while True :
    
    response=requests.get(f'https://www.pararius.com/apartments/amsterdam/page-{page_num}')
    soup=BeautifulSoup(response.text,'html.parser')
    
    with open('appart.csv','a',encoding='utf-8',newline='') as f:
        
        writer=csv.writer(f)


        for ap in soup.find_all('li' , attrs={'class' : 'search-list__item search-list__item--listing'}):
            row.append(ap.find('h2' , attrs={'class' : 'listing-search-item__title'}).get_text().strip())
            row.append(ap.find('div' , attrs={'class' : 'listing-search-item__sub-title'}).get_text().strip())
            row.append(ap.find('li' , attrs={'class' : 'illustrated-features__item illustrated-features__item--surface-area'}).get_text().strip())
            row.append(ap.find('div' , attrs={'class':'listing-search-item__price'}).get_text().strip())
            row.append(ap.find('li' , attrs={'class' : 'illustrated-features__item illustrated-features__item--number-of-rooms'}).get_text().strip())
            try: 
                 row.append(ap.find('ul' , attrs={'class' : 'illustrated-features illustrated-features--compact'}).find('li',attrs={'class':'illustrated-features__item illustrated-features__item--interior'}).get_text())
            except :
                 row.append('UnFurnishered')
            writer.writerow(row)
            row=[]    
                
                
        page=soup.find('div', attrs={'class':'pagination__summary'}).get_text().split('of')
        if page[0].split()[2] in page[1].split()[0] :
            print ('finished')
            break
        else :    
            page_num+=1
    
```
