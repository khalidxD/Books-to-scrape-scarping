```python
import requests
from bs4 import BeautifulSoup
url='https://ocw.mit.edu/courses/6-006-introduction-to-algorithms-spring-2020/resources/lecture-videos/'
response=requests.get(url)
soup=BeautifulSoup(response.text,'html.parser')
```

```python
import csv
row=[]
with open('lect.csv','w',encoding='utf-8',newline='') as f:
    writer=csv.writer(f)
    writer.writerow(['Lecture title','Video URL'])
    
    
with open('lect.csv','a',encoding='utf-8',newline='') as f:
    writer=csv.writer(f)
    for i in soup.find_all('a', attrs={'class':'resource-list-title'}):

        row.append(i.get_text().strip())
        row.append(i.get('href'))
        writer.writerow(row)
        row=[]
```
